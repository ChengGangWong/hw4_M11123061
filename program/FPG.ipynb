```python
#資料處理
dataPath=r'Trade_Data_Normalization_v7.csv'
dataList=[]
import csv
with open(dataPath, "r", newline="") as csvData:
  reader = csv.reader(csvData)
  for r in reader:
     dataList.append(r)

#FP-Growth
itemTimes=[]
for d in dataList:
    for i in d:
        isSave=False
        for t in range(0,len(itemTimes)):
            if(i==itemTimes[t][0]):
                isSave=True
                itemTimes[t][1]=itemTimes[t][1]+1
                break
        if(isSave==False):
            items=[i,1]
            itemTimes.append(items)
times=[]
for i in itemTimes:
    times.append(i[1])
times.sort()
times.reverse()
orderTimes=[]
for t in times:
    isSave=False
    for o in orderTimes:
        if(t==0):
            isSave=True
            break;
    if(isSave==False):
        orderTimes.append(t)
orderItem=[]
for o in orderTimes:
    for i in itemTimes:
        if(o==i[1]):
            orderItem.append(i)
supportInput= input("請輸入支持度:")
print("您輸入的支持度為:"+supportInput)
support=len(dataList)*float(supportInput)
itemInSupport=[]
for o in orderItem:
    if(o[1]>=support):
        itemInSupport.append(o)

orderData=[]
for o in dataList:
    dList=[]
    isHave=False
    for i in itemInSupport:
        for o2 in o:
            if(i[0]==o2):
                isHave=True
                dList.append(o2)
    if(isHave==True and len(o)==len(dList)):
        orderData.append(dList)
orderData2=[]
for i in itemInSupport:
    for o in orderData:
        if(i[0]==o[0]):
            orderData2.append(o)
dataNode=[]
for o in orderData2:
    client=0
    for o2 in o:
        main=client+1
        isAdd=True
        for d in dataNode:
            if(d[0]==client and d[1]==o2 and d[2]==main):
                isAdd=False
        if(isAdd==True):
            data=[client,o2,main]
            dataNode.append(data)
        client=client+1
nowList=[]
pathList=[]
nowNum=0
exectue=0
for d in dataNode:
    exectue=exectue+1
    if(nowNum<d[2]):
        nowList.append(d[1])
        nowNum=d[2]
    else:
        nowNum=d[2]
        pathList.append(nowList)
        tempList=[]
        for n in range(0,nowNum-1):
            tempList.append(nowList[n])
        nowList=tempList
        nowList.append(d[1])
    if(len(dataNode)==exectue):
        pathList.append(nowList)
item= input("請輸入商品類型(PRODUCT_TYPE，若有多項請以「,」分開):")
print("您輸入的商品類型為:"+item)
inputItem=item.split(',')
data=[]
for p in pathList:
    chkTest=0
    for i in inputItem:
        for p2 in p:
            if(p2==i):
                chkTest=chkTest+1
    if(chkTest==len(inputItem)):
        for p2 in p:
            isSave=False
            for d in data:
                if(p2==d):
                    isSave=True
                    break
            if(isSave==False):
                data.append(p2)
dArray=[]
for d in data:
    isSave=False
    for i in inputItem:
        if(d==i):
            isSave=True
    if(isSave==False):
        dArray.append(d)
print(dArray)
```
