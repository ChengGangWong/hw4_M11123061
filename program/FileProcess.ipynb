```python
import openpyxl
wb = openpyxl.load_workbook('TradeData.xlsx')
sheet1 = wb.active
dataList=[]
for r in range(sheet1.max_row):
    for c in range(sheet1.max_column):
        dataList.append(sheet1.cell(r+1,c+1).value)
```
